```python
metros = df[df.CAT_B.str.len() == 3]
totalMetros = metros.groupby(["SIC7_2d","SIC7_3d","TaxYear"])['FTE'].sum().reset_index()
totalMetros = pd.DataFrame(totalMetros[totalMetros.SIC7_2d == 14].set_index(['SIC7_2d','SIC7_3d',"TaxYear"]))
totalMetros.columns = ['All Metros']
totalMetros
```

```python
metros.CAT_B.unique()

for metro in ['CPT', 'EKU', 'ETH', 'JHB', 'TSH']:
    metro_df = metros[metros.CAT_B == metro]
    metro_df = metro_df.groupby(["SIC7_2d","SIC7_3d","TaxYear"])['FTE'].sum().reset_index()
    metro_df = pd.DataFrame(metro_df[metro_df.SIC7_2d == 14].set_index(['SIC7_2d','SIC7_3d',"TaxYear"]))
    metro_df.columns = [metro]
    SA_total = SA_total.join(metro_df)

```
